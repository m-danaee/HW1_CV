## 1. Import Required Libraries

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import os

# Set up matplotlib for displaying images inline for better visualization
plt.rcParams['figure.figsize'] = (10, 6)
```

## 2. Load the Image

```python
# Image path
image_path = 'assets/histogram/0057.png'

# Load the image (default is in BGR format)
img_color = cv.imread(image_path)

# Convert BGR (OpenCV format) to RGB (for proper display in matplotlib)
img_rgb = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)

# Convert to grayscale
img_gray = cv.cvtColor(img_color, cv.COLOR_BGR2GRAY)

print(f"Image loaded successfully!")
```

## 3. Display Original Image

```python
if 'img_gray' in locals():
    plt.imshow(img_gray, cmap='gray')
    plt.title('Original Grayscale Image')
    plt.axis('off')
    plt.show()
else:
    print("Please load an image first.")
```

## 4. Display Original Histogram

```python
if 'img_gray' in locals():
    # Calculate histogram
    hist_original = cv.calcHist([img_gray], [0], None, [256], [0, 256])

    # Plot histogram
    plt.plot(hist_original, color='black')
    plt.title('Histogram of Original Image')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.5)
    plt.show()
else:
    print("Please load an image first.")
```

## 5. Apply Histogram Equalization

```python
if 'img_gray' in locals():
    # Apply histogram equalization
    img_equalized = cv.equalizeHist(img_gray)

    print("Histogram equalization applied successfully!")
else:
    print("Please load an image first.")
```

## 6. Display Equalized Image

Visualize the histogram-equalized image.

```python
if 'img_equalized' in locals():
    plt.imshow(img_equalized, cmap='gray')
    plt.title('Histogram Equalized Image')
    plt.axis('off')
    plt.show()
else:
    print("Please apply histogram equalization first.")
```

## 7. Display Equalized Histogram

```python
if 'img_equalized' in locals():
    # Calculate histogram of equalized image
    hist_equalized = cv.calcHist([img_equalized], [0], None, [256], [0, 256])

    # Plot histogram
    plt.plot(hist_equalized, color='blue')
    plt.title('Histogram of Equalized Image')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.5)
    plt.show()
else:
    print("Please apply histogram equalization first.")
```

## 8. Side-by-Side Comparison

```python
if 'img_gray' in locals() and 'img_equalized' in locals():
    fig, axes = plt.subplots(2, 2)

    # Original image
    axes[0, 0].imshow(img_gray, cmap='gray')
    axes[0, 0].set_title('Original Image', fontsize=14, fontweight='bold')
    axes[0, 0].axis('off')

    # Original histogram
    axes[0, 1].plot(hist_original, color='black')
    axes[0, 1].set_title('Original Histogram', fontsize=14, fontweight='bold')
    axes[0, 1].set_xlabel('Pixel Intensity')
    axes[0, 1].set_ylabel('Frequency')
    axes[0, 1].grid(True, alpha=0.5)

    # Equalized image
    axes[1, 0].imshow(img_equalized, cmap='gray')
    axes[1, 0].set_title('Equalized Image', fontsize=14, fontweight='bold')
    axes[1, 0].axis('off')

    # Equalized histogram
    axes[1, 1].plot(hist_equalized, color='blue')
    axes[1, 1].set_title('Equalized Histogram', fontsize=14, fontweight='bold')
    axes[1, 1].set_xlabel('Pixel Intensity')
    axes[1, 1].set_ylabel('Frequency')
    axes[1, 1].grid(True, alpha=0.5)

    plt.tight_layout()
    plt.show()
else:
    print("Please load an image and apply histogram equalization first.")
```
